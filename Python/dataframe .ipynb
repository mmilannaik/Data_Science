##### Pandas group by
    - group by involves 3 sub-operations
        - splitting the object
        - applying a function
        - combining results
        
    - 3 kinds of group level operations
        - aggregation (mean(), max(), min(), sum(), count())
        - transformation 
        - filter

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 
                     'Kings', 'Kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'], 
            'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2], 
            'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
            'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
df.head()
```

###### aggregation (mean(), max(), min(), sum(), count())

```python
###  Get average of "Points" for each 'Team'
df.groupby(['Team'])['Points'].mean()
```

```python
df.groupby(['Team'])[['Points']].mean()
```

```python
df.groupby(['Team'])['Points'].mean().reset_index()
```

```python
df.groupby(['Team','Year'])[['Points']].mean()
```

```python
### for each get maximum of the rank
df.groupby('Team')['Rank'].max().reset_index()
```

```python
df.groupby('Team')['Rank'].max().reset_index().rename(columns = {'Rank':'Max_Rank'})
```

```python
### example of applying same aggregation on multiple columns
df.groupby('Team')['Points','Rank'].mean()
```

```python
df.groupby('Team')[['Points','Rank']].mean()
```

```python
### example of applying different aggregations on differnt columns
## take avg. on 'Points' but maximum of 'Rank'
df.groupby('Team')[['Points','Rank']].agg({'Points':'mean','Rank':'max'})
```

###### group level transformation 

```python
# # 0 index => 
# 876 - df[df.Team == 'Riders']['Points'].mean() 
# # 1st index  => 
# 789 - df[df.Team == 'Riders']['Points'].mean() 
# # 2nd index =>
# 863 - df[df.Team == 'Devils']['Points'].mean() 
# # 4th index =>
# 741 - df[df.Team == 'Kings']['Points'].mean() 
```

```python
df['var_l'] = df.groupby('Team')['Points'].transform(lambda x : x - np.mean(x))
df.head()
```

```python
### normalisation 
#( x - mean(x))/ (max(x)- min(x))
df['nor_l'] = df.groupby('Team')['Points'].transform(lambda x : (x - np.mean(x))/(np.max(x) - np.min(x)))
df.head()
```
