###### Merging dataframes

```python
>>> import pandas as pd 
>>> left_df = pd.DataFrame({ 'id':[1,2,3,4,5],
          'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
                      'subject_id':['sub1','sub2','sub4','sub6','sub5']}) 
>>> right_df = pd.DataFrame( {'id':[1,2,3,4,5],
          'Name': ['Billy', 'Brian', 'Bran', 'Bryce','Betty'],
                           'subject_id':['sub2','sub4','sub3','sub6','sub5']})
```

    Print left and right
    Merge on id
    Merge on id & subject_id
    Perform left join on key = id 
    Perform right join on key = id
    Perform outer join on key = id
    Perform left join on keys = (id, subject_id)
    Perform right join on keys = (id, subject_id)
    Perform outer join on keys = (id, subject_id)

```python
# pd.merge()
```

T1: customer details => customer_id, name, phone number, address
T2: transactional details => transaction_id, customer_id, amount, date, 
T3: product details =>     transaction_id, product_id, product_details

keys/common columns => logical joining
what type of  join => inner join & outer join (full outer join, left outer join, right outer join)

inner join => select all common records 
left outer join => common records + remaining records from left table
right outer join => common records + remaining records from right table
full outer join => common records + remaining records from both left and right tables

```python
left_df
```

```python
right_df
```

```python
# Merge on id
pd.merge(left_df, right_df, on = ['id'])
```

```python
left_df.merge(right_df, on = 'id')
```

```python
# Merge on id & subject_id
left_df.merge(right_df, on = ['id','subject_id'], how = 'inner')
```

```python
# df1 -> [(1,sub1),(2,sub2),(3,sub4),(4,sub6),(5,sub5)]
# df2 -> [(1,su2),(2,sub4),(3,sub3),(4,sub6),(5,sub5)]
```

```python
# Perform left join on key = id 
pd.merge(left_df, right_df, how = 'left', on = 'id')
```

```python
left_df = left_df.append({'id':11, 'Name':'abc','subject_id':'sub11'},ignore_index = True )
```

```python
left_df
```

```python
pd.merge(left_df, right_df, how = 'left', on = 'id')
```

```python
right_df = right_df.append({'id':12, 'Name':'xyz','subject_id':'sub12'},ignore_index = True )
right_df
```

```python
# Perform right join on key = id
pd.merge(left_df, right_df, on = 'id', how='right')
```

```python
# Perform outer join on key = id
pd.merge(left_df, right_df, on = 'id', how='outer')
```

```python
# Perform left join on keys = (id, subject_id)
pd.merge(left_df, right_df, on = ['id','subject_id'], how='left')
```

```python
# df1 -> [(1,sub1),(2,sub2),(3,sub4),(4,sub6),(5,sub5)]
# df2 -> [(1,su2),(2,sub4),(3,sub3),(4,sub6),(5,sub5)]
```

```python
# Perform right join on keys = (id, subject_id)
pd.merge(left_df, right_df, on = ['id','subject_id'], how='right')
```

```python
pd.merge(right_df, left_df, on = ['id','subject_id'], how='left')
```

```python
# Perform outer join on keys = (id, subject_id)
pd.merge(left_df, right_df, on = ['id','subject_id'], how='outer')
```
